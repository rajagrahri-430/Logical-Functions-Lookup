#Logical Functions & Lookup
Q1. Write a formula to check if the marks in cell are greater than or equal to 40. If yes, return "pass", otherwise "fail".

> Answer
- =IF(B2>=40,"Pass","Fail")

What it does, plainly:

Checks if the value in B2 is 40 or more.

If yes → returns Pass

If no → returns Fail

Q3. rite a nested IF formula:
If marks in B2 are ≥ 90 →'A',≥ 75 →'B', ≥ 50 →'C', otherwise"Fail".
> Answer
- =IF(B2>=90,"A",IF(B2>=75,"B",IF(B2>=50,"C","Fail")))

Logic, straight and clean:

≥ 90 → A

Q4. Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?
>Answer
- Formula (using VLOOKUP):

**=VLOOKUP("E103", A2:C5, 3, FALSE)**

Result: Finance

Why this works:

We search Employee ID E103 in the first column.

Department is in the 3rd column, so column index = 3.

FALSE forces exact match.

Which function and why:

VLOOKUP is used because:

Lookup value is in the first column.

Data is vertical.

Simple left-to-right search.

This is the correct tool for this table structure.

Q5. company has set the following commission structure for its sales team:

> Answer
- `=VLOOKUP(45000, A2:B5, 2, TRUE)`

**Result:** **10%**

Why:

* Sales slabs are in **ascending order**.
* 45,000 lies between **20,000 and 50,000**.
* Excel picks the **nearest lower slab → 20,000 → 10%**.

Function used:

* **VLOOKUP with TRUE (approximate match)**
  Used because commission depends on **range**, not exact value.
